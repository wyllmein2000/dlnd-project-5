## Neural Network Training
### MNIST
Test your GANs architecture on MNIST.  After 2 epochs, the GANs should be able to generate images that look like handwritten digits.  Make sure the loss of the generator is lower than the loss of the discriminator or close to 0.

```python
from gan import *

batch_size = 128
z_dim = 100
learning_rate = 0.001
beta1 = 0.5
epochs = 2

mnist_dataset = helper.Dataset('mnist', glob(os.path.join(data_dir, 'mnist/*.jpg')))


aaa = dcgan(batch_size, learning_rate, z_dim, beta1, epochs)
aaa.check_gpu()
aaa.unit_test()

with tf.Graph().as_default():
    aaa.train(mnist_dataset.get_batches, mnist_dataset.shape, mnist_dataset.image_mode)
```

### CelebA
Run your GANs on CelebA.  It will take around 20 minutes on the average GPU to run one epoch.  You can run the whole epoch or stop when it starts to generate realistic faces.

```python
batch_size = 128
z_dim = 100
learning_rate = 0.001
beta1 = 0.5
epochs = 1

celeba_dataset = helper.Dataset('celeba', glob(os.path.join(data_dir, 'img_align_celeba/*.jpg')))

bbb = dcgan(batch_size, learning_rate, z_dim, beta1, epochs)
    
with tf.Graph().as_default():
    bbb.train(celeba_dataset.get_batches, celeba_dataset.shape, celeba_dataset.image_mode)
```
